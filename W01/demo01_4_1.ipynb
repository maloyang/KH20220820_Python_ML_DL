## 新北市公共自行車租賃系統(YouBike): https://data.gov.tw/dataset/123026

```python
import requests
```

```python
url = "https://data.ntpc.gov.tw/api/datasets/71CD1490-A2DF-4198-BEF1-318479775E8A/csv/file"
```

```python
r = requests.get(url)
```

```python
r
```

```python
decoded_content = r.content.decode('utf-8')
```

```python
print(decoded_content)
```

```python
# 用csv套件解析

import csv
cr = csv.reader(decoded_content.splitlines(), delimiter=',')
data_list = list(cr)
```

```python
print(data_list[0])
```

```python
print(data_list[1])
```

```python
import folium
import geocoder

# 使用 geocoder 取得特定住址的 GPS 座標
location = geocoder.osm('新北市').latlng
print(location)
```

```python
# ['"sno"', 'sna',   'tot',  'sbi', 'sarea', 'mday',      'lat',    'lng',    'ar',              'sareaen', 'snaen', 'aren', 'bemp', 'act']
# ['1001', '大鵬華城', '28',  '9',  '新店區', '20220820143915', '24.99116', '121.53398', '新北市新店區中正路700巷3號', 'Xindian Dist.', 'Dapeng Community', 'No. 3, Lane 700 Chung Cheng Road, Xindian District', '15', '1']
```

```python
# 請在這邊填入你的程式

# TODO: 請在這邊填入你的程式

m = folium.Map(location=location, zoom_start=15)

for item in data_list[1:]:
    try:
        # item[2]=緯度, item[1]=經度, item[0]=名稱
        name = item[1]  #站名
        total = item[2] #可借總數量
        sbi = item[3]  #剩餘數量
        lat = item[6]  #緯度
        lng = item[7]  #經度
        info = '%s 剩餘數量[%s/%s]' %(name, sbi, total)
        folium.Marker([ float(lat), float(lng)],
                      tooltip=info,
                     icon=folium.Icon(color='blue', prefix='fa', icon='fa-bicycle')
                     ).add_to(m)
        
    except Exception as e:
        print(e.args)    
```
